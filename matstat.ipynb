**Функция распределения** \\
\
**Опр.** Функцией распределения случайной велечины $\xi (ω)$ называется функция $F_\xi (x) = P(\omega : \xi (\omega) <= x) = P(\xi <= x)$

 Функция распределения: $F_\xi(x)$ = P($\xi <= x$) = $\sum_{i = 1}^{\lfloor x \rfloor} 1 * P(\xi = i) = \frac{1}{\theta}\sum_{i = 1}^{\lfloor x \rfloor} 1 =\frac{\lfloor x \rfloor}{\theta} $ \\
$$F_{\xi}(x) = \begin{cases} 0, x<1\\ \frac{\lfloor x \rfloor}{\theta}, x \in [1,\theta] \\ 1, x > \theta \end{cases}$$ \\
$\lfloor x \rfloor$ потому что сама случайная величина от 1 до $\theta$, но функция распределения будет определена на всем числовом пространстве \\

**Математическое ожидание** \\
\
**Опр** Математическим ожиданием **дискретной случайной** величины называют сумму произведений всех возможных значений случайной величины на вероятности этих значений \\
$$ M_{\xi} = \sum_{i = 1}^{\theta} i P(\xi = i) = \frac{1}{\theta} * \frac{(\theta+1)\theta}{2} = \frac{\theta + 1}{2} $$ \\

**Дисперсия** \\
\
**Опр.** Дисперсия - это математическое ожидание $(\xi - M_{\xi})^2$, т.е. ценральный момент 2-го порядка случайной величины $\xi$ \\
$D_{\xi} = M(\xi - M_{\xi})^2$ \\\\
$M_{\xi^2} = \frac{1}{\theta} \sum_{i =1}^{\theta} = \frac{\theta(\theta+1)(2\theta+1)}{6} * \frac{1}{\theta} = \frac{(\theta + 1)(2\theta+1)}{6}$ - по формуле суммы квадратов \\
$$D_{\xi} = M_{(\xi -M_{\xi})^2} = M_{\xi^2} - (M_{\xi})^2 =\frac{(\theta + 1)(2\theta+1)}{6} -  (\frac{\theta + 1}{2})^2 = \frac{\theta^2 - 1}{12}$$ \\

\
**Опр.** Квантиль уровня $\gamma$ - это такое значение x случайной величины, при котором ее функция распределения принимает значение $\gamma$ \\
<p align="center">Другими словами это x:

#### Дискретное равномерное I



Дискретное равномерное распределение с параметром $θ$ можно задать с помощью таблицы распределения:
$$
𝒫 :  \bigg(
\begin{array}{ccc}
\text{1}&\text{...}&\text{k}&\text{...}&\text{θ}\\
p_1&\text{...}&p_k&\text{...}&p_θ
\end{array}
\bigg)
$$
Все $p_i$ в таблице одинаковые и равны $\frac{1}{θ}$ \\
Рассмотрим метод обратных функций для моделирования дискретного равномерного распределения на множестве $\{x_1, x_θ\}$ \\
Пусть $s_i = p_1 + \dots + p_i ⇒ s_0 = 0, s_θ = 1$ \\
Рассмотрим $ϕ : (0,1] → \{x_1, x_θ\}$ \\
$$
ϕ(x) = x_i, s_{i-1} < x <= s_{i}, i =1, \dots, θ
$$
Отрезок (0,1] делится на $θ$ на отрезков вида $(s_{i-1}, s_{i}]$. Если наше $\gamma$ окажется в i-ом отрезке, то дискретная равномерная случайная величина примет значение $x_i$. Т.е. простыми словами берем столбики(скачки), на которые поделили отрезок, присваиваем ей индекс такой же как у $s_i$.
У нас есть иcточник непрерывных случайных величин на отрезке [0,1], из него домножением на $θ$ получаем [0, $θ$]. Затем просто оставляем целую часть. Воспользуемся мерой лебега для $ℝ^2$. Вероятность того, что случайная величина попадет в отрезок будет равна отношению длины этого отрезка на длину всего промежутка, отрезки у нас равны, следовательно наша величина будет распредлена дискретно равномерно.

```python
import numpy as np
import random
def gen_discr_uniform( theta = 48):
  gamma = random.uniform(0,1)
  return int(gamma * theta) + 1
gen_discr_uniform()
```

#### Нормальное II

Пусть случайный вектор $X = (X_1, X_2, \dots, X_k)$ имеет абсолютно непрерывное с плотностью $f(x)$ \\
$x = (x_1, \dots, x_k) \in S \subseteq R^k $ \\
$h = (h_1, \dots, h_k) : S → R^k$ - произвольное биективное отображение, частные производные которого непрерывны, якобиан которого не обращается в ноль:
$$J(x) = det \begin{bmatrix}
  \frac{\partial h_1(x)}{\partial x_1} &
    … &
    \frac{\partial h_k(x)}{\partial x_1} \\[1ex] % <-- 1ex more space between rows of matrix
  \dots &
    \dots &
    \dots \\[1ex]
  \frac{\partial h_1(x)}{\partial x_k} &
    … &
    \frac{\partial h_k(x)}{\partial x_k}
\end{bmatrix} $$ \\
Тогда плотность плотность распределения преобразованного случайного вектора $Y= h(X) = (h_1(X), \dots, h_k(X))$ имеет следующий вид:
$$
g(y) = \frac{f(h^{-1}(y))}{J(h^{-1}(y))}, y =(y_1, \dots, y_k) \in h(S) \\
$$
Пусть $X_1, X_2 \sim U(0,1)$ - независимы. Рассмотрим их преобразование:
$$
Y_1 = \sqrt{-2 \ln{X_1}} \sin{(2\pi X_2)} \\
Y_2 = \sqrt{-2 \ln{X_1}} \cos{(2\pi X_2)}
$$
Найдем распределение пары $Y = (Y_1, Y_2)$ \\
$$
y_1 = h_1(x_1, x_2) = \sqrt{-2 \ln{x_1}} \sin{(2\pi x_2)} \\
y_2 = h_2(x_1, x_2) = \sqrt{-2 \ln{x_1}} \cos{(2\pi x_2)} \\
(x_1, x_2) \in (0,1)^2
$$
Тогда $g(y_1, y_2) = \frac{x_1}{2\pi}$ \\
$y_1^2 + y_2^2 = -lnx_1 => x_1 = \exp \{-\frac{1}{2}(y_1^2 + y_2^2) \}$ \\
$$g(y_1, y_2) = \frac{\exp \{-\frac{1}{2}(y_1^2 + y_2^2) \}}{2\pi}$$
Получили, что искомая плотность имеет вид двумерной стандартной нормальной плотности, т.е. его компоненты независимы и распределены $N(0,1)$ \\
Из полученного результата следует, что каждая пара $(U_{2k-1}, U_{2k}), k = 1,2 \dots $ из последовательности независимых равномерно распределенных на интервале [0,1] случайных чисел с помощью преобразования
$$
X_{2k-1} = \sqrt{-2 \ln{U_{2k-1}}} \sin{(2\pi U_{2k})} \\
X_{2k} = \sqrt{-2 \ln{U_{2k-1}}} \cos{(2\pi U_{2k})}
$$
порождает пару $(X_{2k-1}, X_{2k})$ независимых стандартных нормальных чисел. Если мы хотим сгенерировать случайное число с $N(\mu, θ^2)$ воспользуемся, что $X = \mu + θN(0,1)$

```python
import math
def gen_normal_distib( mu = 5.0, theta = 8.0):
  u1 = next(x for x in iter(lambda: random.uniform(0, 1), 0) if x > 0)
  u2 = random.uniform(0,1)
  return mu + theta * math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
gen_normal_distib()
```

#### Дискретное равномерное I

Функция генерации выборок для выбранных случайных величин

```python
def gen_distribution(sample_sizes, k, distr = "uniform"):
  if distr == "uniform":
    return {n: [[gen_discr_uniform() for _ in range(n)] for _ in range(k)] for n in sample_sizes}
  elif distr == "normal":
    return {n: [[gen_normal_distib() for _ in range(n)] for _ in range(k)] for n in sample_sizes}
```

```python
samples_uniform = gen_distribution(sample_sizes, 5, distr = "uniform")
samples_normal = gen_distribution(sample_sizes, 5, distr = 'normal')
```

###  Построение эмпирической функции распределения
****

Теперь займемся графиками \\
Для каждой сгенерированной выборки необходимо построить график эмпирической функции распределения. На одном графике необходимо
отобразить эмпирические функции распределения для каждого из объемов
выборки независимо и график функции распределения случайной величины.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from scipy.special import erf

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#00FF00', '#0000FF', '#00FFFF', '#FFFF00', '#FF00FF']

# Построение эмпирической функции распределения
def empirical_cdf(sample, t_values):
    n = len(sample)
    return np.array([np.sum(sample < t) / n for t in t_values])

# Теоретическая ФР для равномерного распределения
def uniform_cdf(t, theta=48):
    return np.clip(t / theta, 0, 1)

# Теоретическая ФР для нормального распределения
def normal_cdf(t, mu=5.0, sigma=8.0):
    return 0.5 * (1 + erf((t - mu) / (sigma * math.sqrt(2))))

def plot_empirical_cdf(sample_sizes, k, samples_uniform, samples_normal):
    t_values_uniform = np.linspace(0, 49, 48)  # Диапазон значений t для равномерного распределения
    t_values_normal = np.linspace(-20, 40, 600)  # Диапазон значений t для нормального распределения


    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    # Построение графиков для дискретного равномерного распределения
    for i, n in enumerate(sample_sizes):
        sample = np.array(samples_uniform[n][0])  # Берем первую выборку для каждого объема
        emp_cdf_values = empirical_cdf(sample, t_values_uniform)
        ax1.step(t_values_uniform, emp_cdf_values,where='post', color=colors[i % len(colors)], label=f'ЭФР для n={n}')

    # Теоретическая ФР для равномерного распределения
    theo_cdf_values_uniform = uniform_cdf(t_values_uniform)
    ax1.step(t_values_uniform, theo_cdf_values_uniform, where='post', color = 'k', label="Теор ФР равномерное")
    ax1.set_title('Дисркетное равномерное распредление с theta = 48')
    ax1.set_xlabel('x')
    ax1.set_ylabel('F_n(x)')
    ax1.legend()
    ax1.grid(True)

    # Построение графиков для нормального распределения
    for i, n in enumerate(sample_sizes):
        sample = np.array(samples_normal[n][0])  # Берем первую выборку для каждого объема
        emp_cdf_values = empirical_cdf(sample, t_values_normal)
        ax2.plot(t_values_normal, emp_cdf_values, color=colors[i % len(colors)], label=f'ЭФР для n={n}')

    # Теоретическая ФР для нормального распределения
    theo_cdf_values_normal = [normal_cdf(t) for t in t_values_normal]
    ax2.plot(t_values_normal, theo_cdf_values_normal, color = 'k', label="Теор ФР нормальное")
    ax2.set_title('Нормальное распредление с mu =5, sigma = 8')
    ax2.set_xlabel('x')
    ax2.set_ylabel('F_n(x)')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

k = 5  # Количество выборок для каждого объема
plot_empirical_cdf(sample_sizes, k, samples_uniform, samples_normal)
```

Для каждой пары построенных эмпирических $F_n(x)$ и $F_m(x)$, n, m ∈
{5, 10, 100, 200, 400, 600, 800, 1000} необходимо вычислить
$$
D_{m,n} = \sqrt{\frac{nm}{n+m}}sup|F_n(x) -F_m(x) | x \in ℝ
$$

```python
def compute_Smirnov(sample_n, sample_m, t_values):
  F_n = empirical_cdf(sample_n, t_values)
  F_m = empirical_cdf(sample_m, t_values)

  n = len(sample_n)
  m = len(sample_m)

  # Если длины эмпирических функций разные, обрезаем до минимальной длины
  min_len = min(len(F_n), len(F_m))
  F_n = F_n[:min_len]
  F_m = F_m[:min_len]

  sup_diff = np.max(np.abs(F_n - F_m))
  D_n_m = math.sqrt(n*m / (n+m)) * sup_diff
  return D_n_m


def compute_and_save_D_mn_values(sample_sizes, samples, t_values, filename):
    with open(filename, "w") as file:
        for i, n in enumerate(sample_sizes):
            for j, m in enumerate(sample_sizes):
               for idx_n, sample_n in enumerate(samples[n]):
                  for idx_m, sample_m in enumerate(samples[m]):
                      D_mn = compute_Smirnov(sample_n, sample_m, t_values)
                      # Форматируем строку для записи в файл
                      result = f"Выборка n: n{n}.{idx_n}, выборка m: m{m}.{idx_m} Dmn={D_mn}\n"
                      file.write(result)
```

```python
# Функция вероятности для равномерного распределения
def uniform_pdf(t, theta=48):
    return np.ones_like(t) / theta * (0 <= t) * (t <= theta)

# Плотность вероятности для равномерного распределения
def normal_pdf(t, mu=5.0, sigma=8.0):
    return (1 / (sigma * math.sqrt(2 * math.pi))) * np.exp(-0.5 * ((t - mu) / sigma) ** 2)
# Построение полигона частот
def plot_frequency_polygon(sample, bins, ax, label, color):
    hist, bin_edges = np.histogram(sample, bins=bins, density=True)
    # Центр бина указывает среднюю точку интервала, которая лучше отображает положение этих данных
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    ax.plot(bin_centers, hist, marker='o', linestyle='-', color=color, label=label)

def plot_multiple_frequency_polygon(sample_sizes, k=5, distr="uniform", samples=samples_uniform, bins = 20):
    t_values_uniform = np.linspace(0, 48, 49)  # Диапазон для дискретного равномерного распределения
    t_values_normal = np.linspace(-20, 40, 500)  # Диапазон для нормального распределения

    fig, axes = plt.subplots(3, 3, figsize=(21, 18))  # 3 строки и 3 столбца
    axes = axes.flatten()  # Уплощаем для удобного доступа

    colors = ['#00FF00', '#0000FF', '#00FFFF', '#FFFF00', '#FF00FF']
    # Рисуем графики ЭФР
    for idx, n in enumerate(sample_sizes):
        ax = axes[idx]
        for i in range(k):
            sample = np.array(samples[n][i])
            t_values = t_values_uniform if distr == "uniform" else t_values_normal
            # uniform_pdf = uniform_pdf(sample, t_values)
            if distr == "uniform":
              plot_frequency_polygon(sample, bins = bins, ax = ax, label = f"P(x) {i+1}", color=colors[i % len(colors)] )
            elif distr == "normal":
              plot_frequency_polygon(sample, bins = bins, ax = ax, label = f"f(x) {i+1}", color=colors[i % len(colors)] )

        if distr == "uniform":
            ax.plot(t_values_uniform, uniform_pdf(t_values_uniform),
                    color='k', label='Теор. плотность')
        elif distr == "normal":
            ax.plot(t_values_normal, normal_pdf(t_values_normal),
                    color='k', label='Теор. плотность')

        # Заголовки и оформление
        ax.set_title(f'{distr} полигон частот, n={n}')
        ax.set_xlabel('x')
        if distr == "uniform":
           ax.set_ylabel('P(x)')
        elif distr == "normal":
           ax.set_ylabel('f(x)')
        ax.legend()
        ax.grid(True)


    fig.delaxes(axes[-1])# Удаляем последнюю пустую ячейку
    plt.tight_layout()
    plt.show()
```

Из графиков видно, что при увеличении размера выборки, гистограммы частот, эмпирические функции распределения приближаются к функции вероятности и плотности распредления у непрерывного распределения, функции распределения. \\
1. Это иллюстрирует теорему сходимости по вероятности ЭФР \\
$$ \text{Теорема для любого x и любого } ɛ > 0 \text{ при } n → ∞ \\
P\{|\widehat{ F_n(x)}  - F_n(x)| < 𝜀 \} → 1
$$
Т.е  графики иллюстрируют свойство состоятельности ЭФР $\widehat{ F_n(x)}$ как оценки для $F_n(x)$
2. Теорему Гливенко, про то, что масимальное отклонение эмпирической функции распределения от теоретической функции распределения на всей оси с вероятностью 1 будет сколь угодно мало при достаточно большом объеме выборки.
$$
P\{lim_{n\rightarrow ∞} D_n = 0\} = 1
$$
3. Теорему Смирнова для определения гипотезы однородности, что две выборки получены из одного и того же распредления
$$
lim_{n,m → ∞} P\{D_{m,n} <= t\} = K(t)
$$
где $K(t)$  - распредление Колмогорова, для опредления доверительной зоны для теоретической ФР



###    Вычисление выборочных моментов
****

Для каждой сгенерированной выборки необходимо выписать значение выборочного среднего $\overline{X}$ и выборочной дисперсии $\overline{S}^2$ \\
$$
\overline{X} = \frac{1}{n} \sum_{i=1}^n X_i \\
\overline{S}^2 = \frac{1}{n} \sum_{i=1}^n (X_i - \overline{X})^2
$$ \\
Какими свойствами обладают эти оценки? \\
Необходимо сравнить значения полученных оценок с истинными
значениями математического ожидания и дисперсии

```python
theta = 48
true_mean_uniform = (1 + theta) / 2
true_variance_uniform = ((theta - 1)**2) / 12

mu, sigma = 5.0, 8.0
true_mean_normal = mu
true_variance_normal = sigma**2

def sample_statistics(sample):
    mean = sum(sample) / len(sample)  # Вычисляем выборочное среднее
    variance = sum((x - mean) ** 2 for x in sample) / len(sample)  # Вычисляем выборочную дисперсию
    return mean, variance

def calculate_statistics_for_samples_to_file(samples_uniform, distr, filename):
    with open(filename, 'w') as file:
        for n, sample_list in samples_uniform.items():
            file.write(f"\nВыборки для n={n} (распределение: {distr}):\n")

            for i, sample in enumerate(sample_list):
                mean, variance = sample_statistics(sample)

                if distr == 'uniform':
                    true_mean, true_variance = true_mean_uniform, true_variance_uniform
                elif distr== 'normal':
                    true_mean, true_variance = true_mean_normal, true_variance_normal

                file.write(f"  Выборка {i+1}:\n")
                file.write(f"    Выборочное среднее: {mean}, Истинное: {true_mean}\n")
                file.write(f"    Выборочная дисперсия: {variance}, Истинная: {true_variance}\n")

```

```python
filename = "statistics_uniform.txt"
calculate_statistics_for_samples_to_file(samples_uniform, 'uniform', filename)
```

```python
filename = "statistics_normal.txt"
calculate_statistics_for_samples_to_file(samples_normal, 'normal', filename)
```

**Выборочное среднее**
1. Является несмещенной оценкой
$$
M_\overline{X} = M(\frac{1}{n} ∑_{i=1}^n X_i) = \frac{1}{n} ∑_{i=1}^n M_{X_i} = M_{X_1} = M_\xi
$$
2. Состоятельная (при увеличении размера выборки она все ближе и ближе подходит к истинному значению параметра) \\
Для доказательства состоятельности оценки воспользуемся неравенством Чебышева.
$$
P\{|\xi - M_\xi|>= ɛ \} <= \frac{D_\xi}{ɛ^2} \\
P\{|\overline{X} - M_\overline{X}|>= ɛ \} <= \frac{D_\overline{X}}{ɛ^2} \\
D_\overline{X} = \frac{D_\xi}{n} \\
P\{|\xi - M_\xi|>= ɛ \} <= \frac{D_\xi}{ɛ^2} = \frac{D_\xi}{n} → 0, n → ∞
$$
**Выборочная дисперсия**
1. Смещенная оценка \\
Пусть $Y_i = X_i - M_\xi$, $i = 1, \dots, n$ \\
$S^2 = \frac{1}{n} \sum_{i=1}^{n}(Y_i -\overline{Y} )^2 = \frac{1}{n} \sum_{i=1}^{n}Y_i^2 - \overline{Y}^2$ \\
Т.к. $M_{Y_i} = 0, M_{Y_i^2} = \mu_2, M_{Y_i Y_j} = M_{Y_j Y_i} = 0 (i \not= j)$ \\
$$M_{\overline{Y}^2} = \frac{1}{n^2} \sum_{i,j =1}^n M_{Y_i Y_j} = \frac{1}{n^2} \sum_{i,j =1}^n M_{Y_i^2} = \frac{\mu_2}{n} \\
M_{S^2} = \frac{n-1}{n}\mu_2 \not= \mu_2
$$
2. Состоятельная оценка \\
$$
P\{|\overline{S}^2 - M_{\overline{S}^2}|>= ɛ \} <= \frac{D_{\overline{S}^2}}{ɛ^2} \\
D_{\overline{S}^2} = \frac{(n-1)^2}{n^3}(\mu_4 - \frac{n-3}{n-1}\mu_2^2) \\
P\{|\overline{S}^2 - M_{\overline{S}^2}|>= ɛ \} <= \frac{D_{\overline{S}^2}}{ɛ^2}  = \frac{(n-1)^2}{ɛ^2 n^3}(\mu_4 - \frac{n-3}{n-1}\mu_2^2) → 0, n → ∞
$$
