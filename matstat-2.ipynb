**Функция распределения** \
\
**Опр.** Функцией распределения случайной велечины $\xi (ω)$ называется функция $F_\xi (x) = P(\omega : \xi (\omega) <= x) = P(\xi <= x)$

 Функция распределения: $F_\xi(x)$ = P($\xi <= x$) = $\sum_{i = 1}^{\lfloor x \rfloor} 1 * P(\xi = i) = \frac{1}{\theta}\sum_{i = 1}^{\lfloor x \rfloor} 1 =\frac{\lfloor x \rfloor}{\theta} $ \
$$F_{\xi}(x) = \begin{cases} 0, x<1\\ \frac{\lfloor x \rfloor}{\theta}, x \in [1,\theta] \ 1, x > \theta \end{cases}$$ \
$\lfloor x \rfloor$ потому что сама случайная величина от 1 до $\theta$, но функция распределения будет определена на всем числовом пространстве

**Математическое ожидание** \
\
**Опр** Математическим ожиданием **дискретной случайной** величины называют сумму произведений всех возможных значений случайной величины на вероятности этих значений \
$$ M_{\xi} = \sum_{i = 1}^{\theta} i P(\xi = i) = \frac{1}{\theta} * \frac{(\theta+1)\theta}{2} = \frac{\theta + 1}{2} $$

**Опр.** Дисперсия - это математическое ожидание $(\xi - M_{\xi})^2$, т.е. ценральный момент 2-го порядка случайной величины $\xi$ \
$D_{\xi} = M(\xi - M_{\xi})^2$ \
$M_{\xi^2} = \frac{1}{\theta} \sum_{i =1}^{\theta} = \frac{\theta(\theta+1)(2\theta+1)}{6} * \frac{1}{\theta} = \frac{(\theta + 1)(2\theta+1)}{6}$ - по формуле суммы квадратов \
$$D_{\xi} = M_{(\xi -M_{\xi})^2} = M_{\xi^2} - (M_{\xi})^2 =\frac{(\theta + 1)(2\theta+1)}{6} -  (\frac{\theta + 1}{2})^2 = \frac{\theta^2 - 1}{12}$$

#### Дискретное равномерное I



Дискретное равномерное распределение с параметром $θ$ можно задать с помощью таблицы распределения:
$$
𝒫 :  \bigg(
\begin{array}{ccc}
\text{1}&\text{...}&\text{k}&\text{...}&\text{θ}\\
p_1&\text{...}&p_k&\text{...}&p_θ
\end{array}
\bigg)
$$
Все $p_i$ в таблице одинаковые и равны $\frac{1}{θ}$ \
Рассмотрим метод обратных функций для моделирования дискретного равномерного распределения на множестве $\{x_1, x_θ\}$ \
Пусть $s_i = p_1 + \dots + p_i ⇒ s_0 = 0, s_θ = 1$ \
Рассмотрим $ϕ : (0,1] → \{x_1, x_θ\}$ \
$$
ϕ(x) = x_i, s_{i-1} < x <= s_{i}, i =1, \dots, θ
$$
Отрезок (0,1] делится на $θ$ на отрезков вида $(s_{i-1}, s_{i}]$. Если наше $\gamma$ окажется в i-ом отрезке, то дискретная равномерная случайная величина примет значение $x_i$. Т.е. простыми словами берем столбики(скачки), на которые поделили отрезок, присваиваем ей индекс такой же как у $s_i$.
У нас есть иcточник непрерывных случайных величин на отрезке [0,1], из него домножением на $θ$ получаем [0, $θ$]. Затем просто оставляем целую часть. Воспользуемся мерой лебега для $ℝ^2$. Вероятность того, что случайная величина попадет в отрезок будет равна отношению длины этого отрезка на длину всего промежутка, отрезки у нас равны, следовательно наша величина будет распредлена дискретно равномерно.

```python
import numpy as np
import random
def gen_discr_uniform(theta = 48):
  gamma = random.uniform(0,1)
  return int(gamma * theta) + 1
gen_discr_uniform()
```

#### Нормальное II

Пусть случайный вектор $X = (X_1, X_2, \dots, X_k)$ имеет абсолютно непрерывное с плотностью $f(x)$ \\
$x = (x_1, \dots, x_k) \in S \subseteq R^k $ \\
$h = (h_1, \dots, h_k) : S → R^k$ - произвольное биективное отображение, частные производные которого непрерывны, якобиан которого не обращается в ноль:
$$J(x) = det \begin{bmatrix}
  \frac{\partial h_1(x)}{\partial x_1} &
    … &
    \frac{\partial h_k(x)}{\partial x_1} \\[1ex] % <-- 1ex more space between rows of matrix
  \dots &
    \dots &
    \dots \\[1ex]
  \frac{\partial h_1(x)}{\partial x_k} &
    … &
    \frac{\partial h_k(x)}{\partial x_k}
\end{bmatrix} $$ \\
Тогда плотность плотность распределения преобразованного случайного вектора $Y= h(X) = (h_1(X), \dots, h_k(X))$ имеет следующий вид:
$$
g(y) = \frac{f(h^{-1}(y))}{J(h^{-1}(y))}, y =(y_1, \dots, y_k) \in h(S) \\
$$
Пусть $X_1, X_2 \sim U(0,1)$ - независимы. Рассмотрим их преобразование:
$$
Y_1 = \sqrt{-2 \ln{X_1}} \sin{(2\pi X_2)} \\
Y_2 = \sqrt{-2 \ln{X_1}} \cos{(2\pi X_2)}
$$
Найдем распределение пары $Y = (Y_1, Y_2)$ \\
$$
y_1 = h_1(x_1, x_2) = \sqrt{-2 \ln{x_1}} \sin{(2\pi x_2)} \\
y_2 = h_2(x_1, x_2) = \sqrt{-2 \ln{x_1}} \cos{(2\pi x_2)} \\
(x_1, x_2) \in (0,1)^2
$$
Тогда $g(y_1, y_2) = \frac{x_1}{2\pi}$ \\
$y_1^2 + y_2^2 = -lnx_1 => x_1 = \exp \{-\frac{1}{2}(y_1^2 + y_2^2) \}$ \\
$$g(y_1, y_2) = \frac{\exp \{-\frac{1}{2}(y_1^2 + y_2^2) \}}{2\pi}$$
Получили, что искомая плотность имеет вид двумерной стандартной нормальной плотности, т.е. его компоненты независимы и распределены $N(0,1)$ \\
Из полученного результата следует, что каждая пара $(U_{2k-1}, U_{2k}), k = 1,2 \dots $ из последовательности независимых равномерно распределенных на интервале [0,1] случайных чисел с помощью преобразования
$$
X_{2k-1} = \sqrt{-2 \ln{U_{2k-1}}} \sin{(2\pi U_{2k})} \\
X_{2k} = \sqrt{-2 \ln{U_{2k-1}}} \cos{(2\pi U_{2k})}
$$
порождает пару $(X_{2k-1}, X_{2k})$ независимых стандартных нормальных чисел. Если мы хотим сгенерировать случайное число с $N(\mu, θ^2)$ воспользуемся, что $X = \mu + θN(0,1)$

```python
import math
def gen_normal_distib( mu = 5.0, theta = 8.0):
  u1 = next(x for x in iter(lambda: random.uniform(0, 1), 0) if x > 0)
  u2 = random.uniform(0,1)
  return mu + theta * math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
gen_normal_distib()
```

Функция генерации выборок для выбранных случайных величин

```python
def gen_distribution(sample_sizes, k, distr = "uniform"):
  if distr == "uniform":
    return {n: [[gen_discr_uniform() for _ in range(n)] for _ in range(k)] for n in sample_sizes}
  elif distr == "normal":
    return {n: [[gen_normal_distib() for _ in range(n)] for _ in range(k)] for n in sample_sizes}
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from scipy.special import erf

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#00FF00', '#0000FF', '#00FFFF', '#FFFF00', '#FF00FF']

# Построение эмпирической функции распределения
def empirical_cdf(sample, t_values):
    n = len(sample)
    return np.array([np.sum(sample < t) / n for t in t_values])
    
# Теоретическая ФР для равномерного распределения
def uniform_cdf(t, theta=48):
    return np.clip(t / theta, 0, 1)

# Теоретическая ФР для нормального распределения
def normal_cdf(t, mu=5.0, sigma=8.0):
    return 0.5 * (1 + erf((t - mu) / (sigma * math.sqrt(2))))

def plot_empirical_cdf(sample_sizes, k, samples_uniform, samples_normal):
    t_values_uniform = np.linspace(0, 49, 48)  # Диапазон значений t для равномерного распределения
    t_values_normal = np.linspace(-20, 40, 600)  # Диапазон значений t для нормального распределения


    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    # Построение графиков для дискретного равномерного распределения
    for i, n in enumerate(sample_sizes):
        sample = np.array(samples_uniform[n][0])  # Берем первую выборку для каждого объема
        emp_cdf_values = empirical_cdf(sample, t_values_uniform)
        ax1.step(t_values_uniform, emp_cdf_values,where='post', color=colors[i % len(colors)], label=f'ЭФР для n={n}')

    # Теоретическая ФР для равномерного распределения
    theo_cdf_values_uniform = uniform_cdf(t_values_uniform)
    ax1.step(t_values_uniform, theo_cdf_values_uniform, where='post', color = 'k', label="Теор ФР равномерное")
    ax1.set_title('Дисркетное равномерное распредление с theta = 48')
    ax1.set_xlabel('x')
    ax1.set_ylabel('F_n(x)')
    ax1.legend()
    ax1.grid(True)

    # Построение графиков для нормального распределения
    for i, n in enumerate(sample_sizes):
        sample = np.array(samples_normal[n][0])  # Берем первую выборку для каждого объема
        emp_cdf_values = empirical_cdf(sample, t_values_normal)
        ax2.plot(t_values_normal, emp_cdf_values, color=colors[i % len(colors)], label=f'ЭФР для n={n}')

    # Теоретическая ФР для нормального распределения
    theo_cdf_values_normal = [normal_cdf(t) for t in t_values_normal]
    ax2.plot(t_values_normal, theo_cdf_values_normal, color = 'k', label="Теор ФР нормальное")
    ax2.set_title('Нормальное распредление с mu =5, sigma = 8')
    ax2.set_xlabel('x')
    ax2.set_ylabel('F_n(x)')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

k = 5  # Количество выборок для каждого объема
plot_empirical_cdf(sample_sizes, k, samples_uniform, samples_normal)
```

Для каждой пары построенных эмпирических $F_n(x)$ и $F_m(x)$, n, m ∈
{5, 10, 100, 200, 400, 600, 800, 1000} необходимо вычислить
$$
D_{m,n} = \sqrt{\frac{nm}{n+m}}sup|F_n(x) -F_m(x) | x \in ℝ
$$

```python
import pandas as pd

def compute_D_n_m(sample1, sample2):
    # Размеры выборок
    m = len(sample1)
    n = len(sample2)
    
    if m == 0 or n == 0:
        raise ValueError("Обе выборки должны содержать хотя бы один элемент.")
    
    # Сортируем выборки
    sorted_sample1 = np.sort(sample1)
    sorted_sample2 = np.sort(sample2)
    
    # Объединяем уникальные значения из обеих выборок
    all_values = np.unique(np.concatenate((sorted_sample1, sorted_sample2)))
    
    # Вычисляем ECDF для первой выборки
    ecdf1 = np.searchsorted(sorted_sample1, all_values, side='right') / m
    
    # Вычисляем ECDF для второй выборки
    ecdf2 = np.searchsorted(sorted_sample2, all_values, side='right') / n
    
    # Вычисляем максимальное абсолютное отклонение
    D_n_m = np.max(np.abs(ecdf1 - ecdf2))
    
    return D_n_m
def compute_smirnov_matrix(sample_sizes, samples):
    # Инициализируем список для хранения результатов
    results = []
    
    # Итерация по всем комбинациям размеров выборок
    for n_size in sample_sizes:
        for m_size in sample_sizes:
            # Получаем список выборок для текущих размеров n и m
            samples_n = samples[n_size]
            samples_m = samples[m_size]
            
            # Итерация по всем выборкам для текущих n и m
            for idx_n, sample_n in enumerate(samples_n, 1):
                for idx_m, sample_m in enumerate(samples_m, 1):
                    # Вычисляем D_plus, D_minus, D_n_m
                    D_n_m = compute_D_n_m(sample_n, sample_m)
                    
                    # Формируем множитель
                    multiplicator = math.sqrt(n_size*m_size / (n_size+m_size))
                    
                    # Добавляем запись в список результатов
                    results.append({
                        'Выборка_n': idx_n,
                        'Размер_n': n_size,
                        'Выборка_m': idx_m,
                        'Размер_m': m_size,
                        'D_n_m': D_n_m,
                        'multiplicator*D_n_m': multiplicator*D_n_m
                    })

    # Создаём DataFrame из списка результатов
    return pd.DataFrame(results)
```

```python
# Функция вероятности для равномерного распределения
def uniform_pdf(t, theta=48):
    return np.ones_like(t) / theta * (0 <= t) * (t <= theta)

# Плотность вероятности для равномерного распределения
def normal_pdf(t, mu=5.0, sigma=8.0):
    return (1 / (sigma * math.sqrt(2 * math.pi))) * np.exp(-0.5 * ((t - mu) / sigma) ** 2)
# Построение полигона частот
def plot_frequency_polygon(sample, bins, ax, label, color):
    hist, bin_edges = np.histogram(sample, bins=bins, density=True)
    # Центр бина указывает среднюю точку интервала, которая лучше отображает положение этих данных
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    ax.plot(bin_centers, hist, marker='o', linestyle='-', color=color, label=label)

def plot_multiple_frequency_polygon(sample_sizes, k=5, distr="uniform", samples=samples_uniform, bins = 20):
    t_values_uniform = np.linspace(0, 48, 49)  # Диапазон для дискретного равномерного распределения
    t_values_normal = np.linspace(-20, 40, 500)  # Диапазон для нормального распределения

    fig, axes = plt.subplots(3, 3, figsize=(21, 18))  # 3 строки и 3 столбца
    axes = axes.flatten()  # Уплощаем для удобного доступа

    colors = ['#00FF00', '#0000FF', '#00FFFF', '#FFFF00', '#FF00FF']
    # Рисуем графики ЭФР
    for idx, n in enumerate(sample_sizes):
        ax = axes[idx]
        for i in range(k):
            sample = np.array(samples[n][i])
            t_values = t_values_uniform if distr == "uniform" else t_values_normal
            # uniform_pdf = uniform_pdf(sample, t_values)
            if distr == "uniform":
              plot_frequency_polygon(sample, bins = bins, ax = ax, label = f"P(x) {i+1}", color=colors[i % len(colors)] )
            elif distr == "normal":
              plot_frequency_polygon(sample, bins = bins, ax = ax, label = f"f(x) {i+1}", color=colors[i % len(colors)] )

        if distr == "uniform":
            ax.plot(t_values_uniform, uniform_pdf(t_values_uniform),
                    color='k', label='Теор. плотность')
        elif distr == "normal":
            ax.plot(t_values_normal, normal_pdf(t_values_normal),
                    color='k', label='Теор. плотность')

        # Заголовки и оформление
        ax.set_title(f'{distr} полигон частот, n={n}')
        ax.set_xlabel('x')
        if distr == "uniform":
           ax.set_ylabel('P(x)')
        elif distr == "normal":
           ax.set_ylabel('f(x)')
        ax.legend()
        ax.grid(True)


    fig.delaxes(axes[-1])# Удаляем последнюю пустую ячейку
    plt.tight_layout()
    plt.show()
```

Для каждой сгенерированной выборки необходимо выписать значение выборочного среднего $\overline{X}$ и выборочной дисперсии $\overline{S}^2$ \
$$
\overline{X} = \frac{1}{n} \sum_{i=1}^n X_i$$
$$\overline{S}^2 = \frac{1}{n} \sum_{i=1}^n (X_i - \overline{X})^2
$$ 
Какими свойствами обладают эти оценки? \
Необходимо сравнить значения полученных оценок с истинными
значениями математического ожидания и дисперсии

```python
theta = 48
true_mean_uniform = (1 + theta) / 2
true_variance_uniform = ((theta - 1)**2) / 12

mu, sigma = 5.0, 8.0
true_mean_normal = mu
true_variance_normal = sigma**2

def sample_mean(sample):
    return sum(sample) / len(sample)  # Вычисляем выборочное среднее

def sample_variance(sample):
    mean = sample_mean(sample)
    return sum((x - mean) ** 2 for x in sample) / len(sample )# Вычисляем выборочную дисперсию

def calculate_statistics_for_samples_to_ds(samples, distr, stat = 'mean'):
    table_data = {}
    print(f"Выборки {distr} распределения статистики {stat}")
    if distr == 'uniform':
        if stat == 'mean':
            print(f"Истинное значение {stat} распределения {distr}: {true_mean_uniform}")
        elif stat == 'variance':
            print(f"Истинное значение {stat} распределения {distr}: {true_variance_uniform}")
    elif distr == 'normal':
        if stat == 'mean':
            print(f"Истинное значение {stat} распределения {distr}: {true_mean_normal}")
        elif stat == 'variance':
            print(f"Истинное значение {stat} распределения {distr}: {true_variance_normal}")

    for n, sample_list in samples.items():
        row_label = f"Выборка размера {n}"
        table_data[row_label] = []
        for i, sample in enumerate(sample_list):
            if stat == 'mean':
                mean = sample_mean(sample)
                table_data[row_label].append(mean)
            elif stat == 'variance':
                variance = sample_variance(sample)
                table_data[row_label].append(variance)
                
    df = pd.DataFrame.from_dict(table_data, orient='index', columns=[f"Выборка {i+1}" for i in range(5)])
    return df
```

**Выборочное среднее**
1. Является несмещенной оценкой
$$
M_\overline{X} = M(\frac{1}{n} ∑_{i=1}^n X_i) = \frac{1}{n} ∑_{i=1}^n M_{X_i} = M_{X_1} = M_\xi
$$
2. Состоятельная (при увеличении размера выборки она все ближе и ближе подходит к истинному значению параметра) \\
Для доказательства состоятельности оценки воспользуемся неравенством Чебышева.
$$P \{|\xi - M_\xi|>= \varepsilon \} <= \frac{D_\xi}{\varepsilon ^2}
$$
$$P \{|\overline{X} - M_\overline{X}|>= \varepsilon \} <= \frac{D_\overline{X}}{\varepsilon^2}$$
$$D_\overline{X} = \frac{D_\xi}{n}$$
$$P \{|\xi - M_\xi|>= \varepsilon \} <= \frac{D_\xi}{\varepsilon^2} = \frac{D_\xi}{n} → 0, n → \infty$$
**Выборочная дисперсия**
1. Смещенная оценка \
Пусть $Y_i = X_i - M_\xi$, $i = 1, \dots, n$ \\
$S^2 = \frac{1}{n} \sum_{i=1}^{n}(Y_i -\overline{Y} )^2 = \frac{1}{n} \sum_{i=1}^{n}Y_i^2 - \overline{Y}^2$ \\
Т.к. $M_{Y_i} = 0, M_{Y_i^2} = \mu_2, M_{Y_i Y_j} = M_{Y_j Y_i} = 0 (i \not= j)$ \\
$$M_{\overline{Y}^2} = \frac{1}{n^2} \sum_{i,j =1}^n M_{Y_i Y_j} = \frac{1}{n^2} \sum_{i,j =1}^n M_{Y_i^2} = \frac{\mu_2}{n} \\
M_{S^2} = \frac{n-1}{n}\mu_2 \not= \mu_2
$$
2. Состоятельная оценка \
$$
P\{|\overline{S}^2 - M_{\overline{S}^2}|>= \varepsilon \} <= \frac{D_{\overline{S}^2}}{\varepsilon^2} \\
D_{\overline{S}^2} = \frac{(n-1)^2}{n^3}(\mu_4 - \frac{n-3}{n-1}\mu_2^2) \\
P\{|\overline{S}^2 - M_{\overline{S}^2}|>= \varepsilon \} <= \frac{D_{\overline{S}^2}}{\varepsilon^2}  = \frac{(n-1)^2}{\varepsilon^2 n^3}(\mu_4 - \frac{n-3}{n-1}\mu_2^2) → 0, n → ∞
$$

#### Метод моментов

#### Метод максимального правдаподобия
**Дискретное равномерное I**
$$L(X, \theta) = \prod_{i=1}^{n} P(X_i <= t) Ind (X_i \in {1, \dots \theta}) = \frac{1}{\theta ^ n} Ind(X_{(n)} <= \theta)$$
$L(X, \theta)$ монотонно убывает по $\theta$ для $X_{(n)} <= \theta$, а при $\theta = X_{(n)}$ достигает максиммума, таким образом ОМП: $\theta = X_{(n)}$ \
**Нормаьное II**
$$L(X, \theta) = \prod_{i=1}^{n} f(x) = \prod_{i=1}^{n} \frac{1}{θ \sqrt{2\pi}} \cdot \exp \{-\frac{(X_i-\mu)^2}{2θ^2} \} = \frac{1}{θ^n \sqrt{(2\pi)^n}} \cdot \exp \{-\frac{\sum_{i=1}^n(X_i-\mu)^2}{2θ^2} \}$$

$$L'(x, \theta) = \frac{1}{\sqrt{(2\pi)^n}} \cdot (\frac{-n}{\theta^{n+1}} \cdot \exp \{-\frac{\sum_{i=1}^n(X_i-\mu)^2}{2θ^2} \} + \exp \{-\frac{\sum_{i=1}^n(X_i-\mu)^2}{2θ^2} \} \cdot \theta^{-n-3}\sum_{i=1}^n(X_i-\mu)^2 ) = 0$$
$L'(x, \theta) = -n\theta^2 + \sum_{i=1}^n(X_i-\mu)^2 ) = 0$ => $\theta = \sqrt{\frac{1}{n} \sum_{i=1}^n(X_i-\mu)^2 }$

```python
def mm_uniform(sample):
    mean = sample_mean(sample)
    return 2*mean - 1
def mm_normal(sample):
    stat = sample_variance(sample)
    return np.sqrt(stat)# Вычисляем выборочную дисперсию
def omp_normal(sample, mu = 5.0):
    return np.sqrt(sum((x - mu) ** 2 for x in sample) / len(sample ))
# Заметим, что оценка омп совпадает с мм 
def omp_uniform(sample):
    theta = np.max(sample)
    return theta

def df_omp_and_mm_stats(samples, distr, true_value):
    table_data = {}
    for n, sample_list in samples.items():
        for i, sample in enumerate(sample_list):
            row_label = f"Выборка размером {n} номер {i+1}"
            table_data[row_label] = []
            # ОМП
            if distr == 'uniform':
                omp = omp_uniform(sample)
                mm = mm_uniform(sample)
                table_data[row_label].append(omp)
                table_data[row_label].append(mm)
            elif distr == 'normal':
                omp = omp_normal(sample)
                mm = mm_normal(sample)
                table_data[row_label].append(omp)
                table_data[row_label].append(mm)
            
                
    df = pd.DataFrame.from_dict(table_data, orient='index', columns=[f"ОМП {true_value}", f"MM {true_value}"])
    return df
```

### Поиск оптимальных оценок

**Теорема(Критерий факторизации)** Для того чтобы статистика T(X) была достаточной для параметра $\theta$, необходимо и достаточно, чтобы функция правдоподобия $L(x, \theta)$  выборки имела вид:
$$L(x, \theta) = g(T(x), \theta)h(x)$$ 
гдеб g и h - неотрицательные функции и h - не зависит от $\theta$ \
**Теорема(Рао - Блекуэлла - Колмогорова)** Оптимальная оценка, если она и существует, явялется функцией от достаточной статистики \
**Опр** Достаточную статистику называют полной, если для функции $\phi(T)$ из того, что 
$$E_\theta \phi(T) = 0$$ для любого $\theta$ следует $\phi(t) \equiv 0$ на всем множестве значений статистики T \
**Теорема** Если существует полная достаточная статистика, то всякая функция от нее является оптимальной оценкой своего математического ожидания
**Критерий Рао-Крамера**
$$T(X) - \tau(\theta) = \alpha(\theta)V(X,\theta)$$, где $\alpha(\theta)$ -некоторая функция от $\theta$

#### Дискретное равномерное
$$L(x,\theta) = \frac{1}{\theta ^ n} Ind(X_{(n)} <= \theta)$$
По критерию факторизации $X_{(n)}$ - достаточная статистика \
**Полнота** \
$P(X_{(n)} <= t) = (\frac{t}{\theta})^n$ => $P(X_{(n)} = t) = \frac{t^n - (t-1)^n}{\theta^n}$
$$E_\theta \phi(X_{(n)}) = \sum_{t=1}^\theta \phi(t) \cdot \frac{t^n - (t-1)^n}{\theta^n} = 0 $$ 
Домножим на $\theta^n$ 
$$\sum_{t=1}^\theta \phi(t) \cdot (t^n - (t-1)^n) = 0 $$
**Докажем по индукции** \
**База** $\theta$ = 1 
$$ \phi(1) \cdot (1^n - (0)^n) = 0 => \phi(1) = 0$$
**Шаг индукции** Пусть верно для $\theta$ < k. Докажем для $\theta = k$ \
Рассмотрим сумму
$$\sum_{t=1}^k \phi(t) \cdot (t^n - (t-1)^n) = 0 $$
Подставляя $\phi(t)=0$ для всех $t<k$, остаётся лишь один слагаемый:
$$ \phi(k) \cdot (k^n - (k-1)^n) = 0$$
Т.к $(k^n - (k-1)^n) > 0$ => $\phi(k) = 0$ \
Следовательно $\phi(X_{(n)}) \equiv 0$ => $X_{(n)}$ - полная достаточная статистика \
**Оптимальность** \
Рассмотрим $E_{\theta} H(X_{(n)}) = \theta$ 
$$S(\theta) = \sum_{t=1}^\theta H(t) \cdot (t^n - (t-1)^n) = \theta^{n+1}$$
Тогда 
$$S(\theta) - S(\theta - 1) = H(\theta) \cdot (\theta^n - (\theta-1)^n) = \theta^{n+1} - (\theta-1)^{n+1}$$
Следовательно
$$H(X_{(n)}) = \frac{X_{(n)}^{n+1} - (X_{(n)}-1)^{n+1}}{X_{(n)}^n - (X_{(n)} - 1)^n}$$ оптимальная оценка $\theta$$$$$

#### Нормальное II
Нормальное II принадлежит сеймейству экспоненциальных моделей, ее плотность можно записать в виде
$$f(x) = \frac{1}{θ \sqrt{2\pi}} \cdot \exp \{-\frac{(x-\mu)^2}{2θ^2} \} = \exp \{ ln\frac{1}{θ \sqrt{2\pi}} -\frac{(x-\mu)^2}{2θ^2} \} = \exp \{ - ln(\theta) - ln(\sqrt{2\pi}) - \frac{(x-\mu)^2}{2θ^2}\}$$
* $A(\theta) = -\frac{1}{2\theta^2}$
* $B(x) = (x-\mu)^2$
* $C(\theta) = -ln(\theta)$
* $D(x) = -ln(\sqrt(2\pi))$

$$\tau (\theta) = -\frac{C'(\theta)}{A'(\theta)} = -\frac{\theta^{-3}}{-\frac{1}{\theta}} = \theta^2$$
$$T(X) = \frac{1}{n} \sum_{i=1}^{n}B(X_i) = \frac{1}{n} \sum_{i=1}^{n} (X_i-\mu)^2$$

По критерию Рао-Крамера статистика T(X) - эффективная, а значит оптимальная, для параметрической функции $\tau(\theta). Заметим так же, что данная оценка является достаточной и полной(экспоненциалная модель)$

$$f_y(y) = f_x(\mu + y\theta)|(\mu + y\theta)_y'| = \frac{1}{θ \sqrt{2\pi}} \cdot \exp \{-\frac{(\mu + y\theta-\mu)^2}{2θ^2} \} \cdot \theta = \frac{1}{\sqrt{2\pi}} \cdot \exp \{-\frac{y^2}{2} \}$$
Получается, что Y ~ N(0,1)
* 2 шаг \
Но мы ведь знаем интересную связь с $\chi^2$
$$\frac{\sum_{i=1}^n (X_i - \mu)^2}{\theta^2} \sim \chi_n^2 \implies \sqrt{\frac{\sum_{i=1}^n (X_i - \mu)^2}{\theta^2}} \sim \chi_n$$
Как известно $E_\xi = \sqrt{2}\frac{\Gamma(\frac{n+1}{2})}{\Gamma(\frac{n}{2})}, \xi \sim \chi_n$
* 3 шаг \
$$\sqrt{T(x)} = \sqrt{\frac{\sum_{i=1}^n (X_i - \mu)^2}{n^2}} = \theta \frac{Y}{\sqrt{n}}$$, где
$Y = \sqrt{\frac{\sum_{i=1}^n (X_i - \mu)^2}{\theta^2}} \sim \chi_n$

Тогда наша оптимальная, т.к. построена от полной достаточной, оценка есть \
$$\theta = \sqrt{T(x)} \frac{\sqrt{n}\frac{\Gamma(\frac{n}{2})}{\Gamma(\frac{n+1}{2})}}{\sqrt{2}}= \sqrt{\frac{\sum_{i=1}^n (X_i - \mu)^2}{2}} \cdot \frac{\Gamma\left(\frac{n}{2}\right)}{\Gamma\left(\frac{n+1}{2}\right)}$$

```python
from scipy.special import gammaln

def uniform_optimal_stat(sample): 
    n = len(sample)
    X_n = np.max(sample)
    # Рассчёт r
    r = X_n / (X_n - 1)
    
    # Вычисляем rn = r^n и rnp1 = r^(n+1) с помощью exp и log
    # rn - 1 = exp(n*log(r)) - 1 = expm1(n*log(r))
    rn_minus_1 = math.expm1(n*math.log(r))
    rnp1_minus_1 = math.expm1((n+1)*math.log(r))
    
    H_X_n = (X_n - 1) * (rnp1_minus_1 / rn_minus_1)
    return H_X_n

def normal_optimal_stat(t, mu=5):
    sample = np.array(t)
    n = len(sample)
    
    # Вычисляем сумму квадратов отклонений от mu
    sum_squared_diff = sum((x - mu) ** 2 for x in sample)
    
    # Используем логарифм гамма для предотвращения переполнения
    log_gamma_ratio = gammaln(n * 0.5) - gammaln((n + 1) * 0.5)
    
    # Рассчитываем статистику
    statistic = np.sqrt(sum_squared_diff * 0.5 ) * np.exp(log_gamma_ratio)
    
    return statistic

def df_umvue_stats(samples, distr, true_value):
    table_data = {}
    for n, sample_list in samples.items():
        for i, sample in enumerate(sample_list):
            row_label = f"Выборка размером {n} номер {i+1}"
            table_data[row_label] = []
            # Оптимальная оценка
            if distr == 'uniform':
                umvue = uniform_optimal_stat(sample)
                table_data[row_label].append(umvue)
            elif distr == 'normal':
                umvue = normal_optimal_stat(sample)
                table_data[row_label].append(umvue)
                
    df = pd.DataFrame.from_dict(table_data, orient='index', columns=[f"Истинное значение оптимальной оцеки {true_value}"])
    return df
```

**Объяснение преобразований для избежания переполнения**

Изначально мы имели выражение:
$$
H(X_{(n)}) = \frac{X_{(n)}^{n+1} - (X_{(n)} -1)^{n+1}}{X_{(n)}^{n} - (X_{(n)} -1)^{n}}.
$$

При больших значениях $X_{(n)}$ и n прямое вычисление степеней может привести к численному переполнению или большой потере точности.

Чтобы этого избежать, выполним следующие преобразования:

1. Введём новую переменную:
   $$
   r = \frac{X_{(n)}}{X_n - 1}.
   $$
   
   Отсюда:
   $$
   X_{(n)} = (X_{(n)} - 1) r.
   $$

2. Перепишем $H(X_{(n)})$ через r. Заметим, что:
   $$
   X_{(n)}^{n+1} - (X_{(n)} -1)^{n+1} = (X_{(n)} - 1)^{n+1}(r^{n+1}-1),
   $$
   $$
   X_{(n)}^{n} - (X_{(n)} -1)^{n} = (X_{(n)} - 1)^{n}(r^{n}-1).
   $$

   После сокращения на $(X_{(n)} - 1)^n$ получаем:
   $$
   H(X_{(n)}) = (X_{(n)} - 1) \frac{r^{n+1}-1}{r^{n}-1}.
   $$

3. Теперь нам нужно вычислить $r^n - 1$ и $r^{n+1} - 1$. При больших $n$ значения $r^n$ могут быть трудны для точного вычисления напрямую. Вместо этого мы используем логарифмы и функцию `math.expm1`, которая вычисляет $\exp(x)-1$ с улучшенной точностью для малых $x$.

   Поскольку:
   $$
   r^n - 1 = \exp(n \ln(r)) - 1 = \mathrm{expm1}(n \ln(r)),
   $$
   $$
   r^{n+1} - 1 = \exp((n+1)\ln(r)) - 1 = \mathrm{expm1}((n+1)\ln(r)).
   $$

   Таким образом, мы избежим непосредственного возведения больших чисел в большие степени, используя логарифмы и `expm1`, улучшив численную стабильность и избежав переполнения.

В итоге, финальная формула:
$$
H(X_{(n)}) = (X_{(n)} - 1) \frac{\exp((n+1)\ln(r))-1}{\exp(n\ln(r))-1} = (X_n - 1) \frac{\mathrm{expm1}((n+1)\ln(r))}{\mathrm{expm1}(n\ln(r))}.
$$

**Критерий Колмогорова–Смирнова** \
Классическая статистика критерия Колмогорова–Смирнова для простой гипотезы о полностью известном распределении определяется как:

$$
D_n = \max(D_n^+, D_n^-)
$$

где

$$
D_n^+ = \max_{1 \leq i \leq n} \left( \frac{i}{n} - F(X_{(i)}) \right),
$$

$$
D_n^- = \max_{1 \leq i \leq n} \left( F(X_{(i)}) - \frac{i - 1}{n} \right).
$$

Здесь $X_{(1)} \leq X_{(2)} \leq \cdots \leq X_{(n)}$ – упорядоченная статистика выборки, а $F(x)$ – теоретическая функция распределения. \
Для выборок n<20 будем для улучшения сходимости к предельному распределению Колмогорова будем использовать Критерий Колмогорова–Смирнова с поправкой Большева
$$
S = \frac{6nD_n + 1}{6\sqrt{n}}.
$$
Сам критерий формулируется следующим образом:
Если n>=20 и при выбранном уровне значимости $\alpha$ число $\lambda_\alpha$ определено отношением $K(\lambda_\alpha) = 1 - \alpha$, то
$$
H_0 \text{ отвергается } \Leftrightarrow \sqrt{n}D_n >= \lambda_\alpha
$$

```python
from scipy.optimize import bisect
from scipy.stats import norm
alpha_levels = [0.01, 0.05, 0.1] # проверять будем для этих уровний значимости

# def normal_cdf(t, mu=5.0, sigma=8.0):
#     return 0.5 * (1 + erf((t - mu) / (sigma * math.sqrt(2))))

def kolmogorov_smirnov_stat_bolshev(sample, mu=5.0, sigma=8.0):
    # Сортируем выборку
    sorted_sample = np.sort(sample)
    n = len(sample)
    F = lambda x: norm.cdf((x - mu) / sigma)    
    # Вычисляем D_n^+ и D_n^-
    D_plus = np.max([(i+1)/n - F(x) for i, x in enumerate(sorted_sample)])
    D_minus = np.max([F(x) - (i)/n for i, x in enumerate(sorted_sample)])
    D_n = max(D_plus, D_minus)
    
    # Применяем поправку Большева
    S = (6*n*D_n + 1) / (6*np.sqrt(n))
    
    return D_n, S

# Функция K(lambda) для асимптотического распределения Колмогорова
def kolmogorov_distribution(lmbda):
    sum_series = np.sum([(-1)**(k-1) * np.exp(-2 * (k**2) * (lmbda**2)) for k in range(1, 10000)])
    return 1 - 2 * sum_series

# Уравнение для поиска lambda_alpha: K(lambda) = 1 - alpha
def equation(lmbda, alpha):
    return kolmogorov_distribution(lmbda) - (1 - alpha)

def lambda_alpha(alpha):
    # Численное решение методом бисекции [0,2] 
    # потому что для уровней значимости, применяемых на практике квантиль обычно не превышает 2
    return round(bisect(equation, 0, 2, args=(alpha,)), 2)

def ks_criterion(samples, alpha_levels):
    # Создаем пустой список для хранения строк данных
    rows = []

    for n, sample_list in samples.items():
        for i, sample in enumerate(sample_list):
            row_label = f"Выборка размером {n} номер {i + 1}"
            
            for alpha in alpha_levels:
                l_alpha = lambda_alpha(alpha) 
                D_n, S = kolmogorov_smirnov_stat_bolshev(sample)  # Предполагается, что функция определена

                # Определяем решение в зависимости от условий
                if n >= 20:
                    decision = 'Отвергнута' if np.sqrt(n) * D_n >= l_alpha else 'Принята'
                else:
                    decision = 'Отвергнута' if S >= l_alpha else 'Принята'

                # Добавляем строку данных
                rows.append({
                    'Выборка': row_label,
                    'Уровень значимости alpha': alpha,
                    'Лямбда alpha': l_alpha,
                    'n**0.5 * D_n': np.sqrt(n) * D_n,
                    'S': S,
                    'Решение': decision
                })

    # Создаем DataFrame из списка строк
    df = pd.DataFrame(rows)
    return df
```

Для применения критерия $ \chi^2 $ необходимо сгруппировать данные по интервалам.
Для того чтобы разбивать на различные интервалы будем использовать различные интервалы: \
**Формула Стерджесса**
$$
k = 1 + \lfloor \log_2(n) \rfloor.
$$
**Правило Скотта**:
  $$
  h = \frac{3.49 \cdot s}{n^{1/3}}, \quad k = \lceil \frac{\text{range}(x_1, \dots, x_n)}{h} \rceil
  $$
**Правило Фридмана-Дьякониса**:
  $$
  h = \frac{2 \cdot \text{IQR}(x_1, \dots, x_n)}{n^{1/3}}, \quad k = \lceil \frac{\text{range(sample)}}{h} \rceil
  $$

Будем использовать равномерное разбиение диапазона, а так же попробуем строить интервалы по квантилям теоретического распределения. Если интервалы построены по квантилям теоретического распределения, то каждый интервал будет иметь одинаковую теоретическую вероятность попадания в него наблюдений, то есть каждый интервал будет содержать примерно одинаковую долю всех данных.

Классический критерий согласия $\chi^2$ : Пусть объем выборки n и наблюдяющиеся значения вектора частот $\nu = (\nu_1, \nu_2, \dots, \nu_N)$ удовлетворяет условиям n>=50, $\nu_j>=5$ для любого j: тогда при заданном уровне значимости:
$$
H_0 \text{ отвергается } \Leftrightarrow \{\dot{\chi_n^2} > \chi_{1-\alpha, N-1}^2\}\}
$$

```python
import numpy as np
from scipy.stats import norm, chi2, iqr

def intervals_calc(sample):
    n = len(sample)
    sample_sorted = np.sort(sample)

    k_sturge = 1 + math.ceil(np.log2(n)) 

    h_scott = 3.49 * np.std(sample) / n**(1/3)
    k_scott = int(np.ceil((sample_sorted[-1] - sample_sorted[0]) / h_scott))

    h_freedman_diaconis = 2 * iqr(sample) / n**(1/3)
    k_freedman_diaconis = int(np.ceil((sample_sorted[-1] - sample_sorted[0]) / h_freedman_diaconis))
    return k_sturge, k_scott, k_freedman_diaconis

def chi_square_test_normal(sample, mu=5.0, sigma=8.0, alpha=0.05, k_initial = 3, mode='quantile'):
    n = len(sample)
    # Определим начальное число интервалов
    k = k_initial
    sample_sorted = np.sort(sample)

    if mode == 'quantile':
        # Генерируем квантильные точки для нормального распределения N(mu, sigma^2)
        u = np.linspace(0, 1, k + 1)
        z_quantiles = norm.ppf(u)  # квантили стандартной нормали
        bins = mu + sigma * z_quantiles
    elif mode == 'uniform':
        # Определим границы интервалов равномерно по диапазону данных
        data_min, data_max = sample_sorted[0], sample_sorted[-1]
        bins = np.linspace(data_min, data_max, k + 1)
    else:
        raise ValueError("Invalid mode. Choose 'quantile' or 'uniform'.")
    
    # Функция распределения для N(mu, sigma^2)
    F = lambda x: norm.cdf((x - mu) / sigma)
    
    # Вычисляем теоретические вероятности для каждого интервала
    p = []
    for j in range(k):
        a_j, a_j1 = bins[j], bins[j + 1]
        p_j = F(a_j1) - F(a_j)
        p.append(p_j)
    p = np.array(p)
    
    # Подсчитываем наблюдаемые частоты по интервалам
    freq, _ = np.histogram(sample, bins=bins)
    
    # Объединяем интервалы, где ожидаемая частота меньше 5
    while True:
        # Вычисляем ожидаемые частоты
        expected = n * p
        
        # Находим интервалы с ожидаемой частотой <5
        low_freq = np.where(expected < 5)[0]
        
        if len(low_freq) == 0:
            break  # Все интервалы удовлетворяют условию
        if k <=2: # Если вдруг выборку не удается поделить хотя бы пополам
            break
        
        # Выбираем первый интервал с малой частотой для объединения
        idx = low_freq[0]
        
        if idx == 0:
            # Объединяем с правым соседним
            bins = np.delete(bins, 1)  # удаляем правую границу
        elif idx == len(p) - 1:
            # Объединяем с левым соседним
            bins = np.delete(bins, idx)  # удаляем текущую границу
        else:
            # Объединяем с соседним интервалом, который имеет меньшую ожидаемую частоту
            if expected[idx - 1] < expected[idx + 1]:
                # Объединяем с левым
                bins = np.delete(bins, idx)
            else:
                # Объединяем с правым
                bins = np.delete(bins, idx + 1)
        
        # Пересчитываем число интервалов
        k -= 1
        
        # Пересчитываем теоретические вероятности
        p = []
        for j in range(k):
            a_j, a_j1 = bins[j], bins[j + 1]
            p_j = F(a_j1) - F(a_j)
            p.append(p_j)
        p = np.array(p)
        
        # Пересчитываем наблюдаемые частоты
        freq, _ = np.histogram(sample, bins=bins)

    # Вычислим статистику хи-квадрат
    chi2_stat = np.sum((freq - n*p)**2 / (n*p))

    # Число степеней свободы для простой гипотезы:
    # k интервалов, оцениваемых параметров 0, значит df = k - 1
    df = k - 1

    # Критическое значение
    chi2_crit = chi2.ppf(1 - alpha, df)

    # Решение
    decision = (chi2_stat > chi2_crit)
    
    return decision, chi2_stat, chi2_crit, k, np.sum(p), bins

def chi_criterion(samples, alpha_levels, mu=5.0, sigma=8.0):
    # Определим правила разбиения
    binning_rules = ['Стерджес', 'Скотт', 'Фридмена-Дайкозиса']
    modes = ['uniform', 'quantile']

    # Создаем пустой список для хранения строк данных
    rows = []

    for n, sample_list in samples.items():
        for i, sample in enumerate(sample_list, 1):
            sample_label = f"Выборка размером {n} номер {i}"
            # Рассчитаем k по разным правилам
            k_sturge, k_scott, k_fd = intervals_calc(sample)
            k_dict = {
                'Стерджес': k_sturge,
                'Скотт': k_scott,
                'Фридмена-Дайкозиса': k_fd
            }
            for alpha in alpha_levels:
                # Инициализируем счетчик решений "Отвергнута"
                reject_count = 0
                # Инициализируем словарь для хранения результатов всех комбинаций
                test_results = {}
                for rule, k in k_dict.items():
                    for mode in modes:
                        # Выполняем тест
                        decision, chi2_stat, chi2_crit, k_final = chi_square_test_normal(
                            sample, mu=mu, sigma=sigma, alpha=alpha, k_initial=k, mode=mode
                        )
                        # Формируем ключи
                        k_key = f'Количество интервалов {rule} mode = \'{mode}\''
                        chi2_key = f'Xи-квадрат {rule} mode = \'{mode}\''
                        crit_key = f'Критическое значение {rule} mode = \'{mode}\''
                        decision_key = f'Решение {rule} mode = \'{mode}\''

                        # Сохраняем результаты
                        test_results[k_key] = k_final
                        test_results[chi2_key] = round(chi2_stat, 4)
                        test_results[crit_key] = round(chi2_crit, 4)
                        test_results[decision_key] = 'Отвергнута' if decision else 'Принята'

                        # Считаем количество отказов
                        if decision:
                            reject_count += 1

                # Общее решение: отвергнута, если >=3 из 6 решений
                overall_decision = 'Отвергнута' if reject_count >=3 else 'Принята'

                # Создаем строку для DataFrame
                row = {
                    'Выборка': sample_label,
                    'Уровень значимости': alpha,
                    'Решение': overall_decision
                }
                # Добавляем результаты всех комбинаций
                row.update(test_results)
                # Добавляем строку в список
                rows.append(row)

    # Создаем DataFrame из списка строк
    df = pd.DataFrame(rows)

    # Определим порядок колонок
    columns_order = ['Выборка', 'Уровень значимости']

    # Добавляем для каждого правила и режима:
    # - Количество интервалов
    # - Xи-квадрат
    # - Критическое значение
    # - Решение
    binning_rules = ['Стерджес', 'Скотт', 'Фридмена-Дайкозиса']
    modes = ['uniform', 'quantile']
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Количество интервалов {rule} mode = \'{mode}\'')
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Xи-квадрат {rule} mode = \'{mode}\'')
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Критическое значение {rule} mode = \'{mode}\'')
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Решение {rule} mode = \'{mode}\'')

    # Добавляем общее решение
    columns_order.append('Решение')

    # Проверим, что все колонки есть
    existing_columns = df.columns.tolist()
    missing_columns = [col for col in columns_order if col not in existing_columns]
    if missing_columns:
        print(f"Внимание: следующие колонки отсутствуют и не будут добавлены в порядок: {missing_columns}")

    # Реорганизуем колонки, добавляя только те, что есть
    columns_order_filtered = [col for col in columns_order if col in existing_columns]
    df = df[columns_order_filtered]

    return df
```

**Критерий согласия Хи-квадрат**

**Сложные гипотезы** \
$$
H_0 = \{ \text{Выборка } \{X_1, X_2, \dots, X_n\} \text{ взята из } N(\mu = 5.0, \sigma^2)\}
$$
$$
H_1 = \{ \text{Выборка } \{X_1, X_2, \dots, X_n\} \text{ взята из } N(\mu = 5.0, \sigma^2) \}
$$

Критерий согласия для хи-квадрат расчитывается на основании утверждения \
**Теорема**
Пусть функции $p_j(\theta)$ для $j = 1, \dots, N$ удовлетворяют следующим условиям:

1. $\sum_{j=1}^N p_j(\theta) = 1$ для любого $\theta \in \Theta$.

2. $p_j(\theta) \geq c > 0$ для любого $j$ и существуют непрерывные производные  
   $\frac{\partial p_j(\theta)}{\partial \theta_k}$ и $\frac{\partial^2 p_j(\theta)}{\partial \theta_k \partial \theta_l}$,  
   где $k, l = 1, \dots, r$.

3. Матрица $\left\|\frac{\partial p_j(\theta)}{\partial \theta_k}\right\|$ размером $N \times r$ имеет ранг $r$ для всех $\theta \in \Theta$.

Тогда распределение $\hat{X}^2_n$ при $H_0$ аппроксимируется $\chi^2(N - 1 - r)$.

Следовательно, для сложной гипотезы $H_0$ мы отвергаем $H_0$, если:

$$
\hat{X}^2_n > \chi^2_{1-\alpha, N-1-r}
$$

## Дискретное равномерное I
**Простые гипотезы** \
$H_0 = \{ \text{Выборка } \{X_1, X_2, \dots, X_n\} \text{ взята из } N(\mu = 5.0, \theta^2 = 64)\}$ \
$H_1 = \{  \text{Выборка } \{X_1, X_2, \dots, X_n\} \text{ не является } N(\mu = 5.0, \theta^2 = 64)\}$

Для дискртеного равномерного неприменим критерий согласия Колмогорова-смирнова, для него можно проверить только хи-квадрат
Здесь все проще, мы будем просто равномерно делить наши интервалы и следить за тем, чтобы частота была больше или равна по возможности 5

```python
def merge_intervals(observed, expected, bin_edges, min_expected=5):
    merged_observed = []
    merged_expected = []
    merged_bin_edges = [bin_edges[0]]
    
    current_obs = observed[0]
    current_exp = expected[0]
    
    for i in range(1, len(observed)):
        if current_exp < min_expected:
            # Объединяем текущий интервал с следующим
            current_obs += observed[i]
            current_exp += expected[i]
        else:
            # Сохраняем текущий интервал
            merged_observed.append(current_obs)
            merged_expected.append(current_exp)
            merged_bin_edges.append(bin_edges[i])
            current_obs = observed[i]
            current_exp = expected[i]
    
    # Добавляем последний интервал
    merged_observed.append(current_obs)
    merged_expected.append(current_exp)
    merged_bin_edges.append(bin_edges[-1])
    
    # Проверяем снова, если после первого прохода все E_j >=5
    # Если нет, повторяем процесс
    if np.any(np.array(merged_expected) < min_expected) and len(merged_observed) >1:
        return merge_intervals(merged_observed, merged_expected, merged_bin_edges, min_expected)
    
    return np.array(merged_observed), np.array(merged_expected), np.array(merged_bin_edges)


def chi_square_test_uniform_range(sample, tetha=48, alpha=0.05, k_initial=10, min_expected=5):
    n = len(sample)
    sample_sorted = np.sort(np.array(sample))
    # Начальное число групп
    k = k_initial
    
    bin_edges = np.linspace(1, tetha +1, k+1)
    bin_edges = np.floor(bin_edges)
    bin_edges = np.unique(bin_edges)

    # Функция для вычисления ожидаемых вероятностей
    def expected_probs(bins):
        return np.diff(bins)/tetha
    
    p = expected_probs(bin_edges)
    
    # Подсчитываем наблюдаемые частоты
    freq, _ = np.histogram(sample, bins=bin_edges, range=(1, tetha+1))
    
    expected = n * p
    
    # Для больших выборок проверяем, что все E_j >= min_expected
    if n >=50:
        if np.any(expected < min_expected):
            # Объединяем интервалы, пока все E_j >= min_expected
            freq, expected, bin_edges = merge_intervals(freq, expected, bin_edges, min_expected)
            k = len(freq)
    
    # Вычисляем статистику хи-квадрат
    chi2_stat = np.sum((freq - expected) ** 2 / expected)

    # Степени свободы: число интервалов - 1 (параметр theta известен, r=0)
    df = k - 1

    # Критическое значение хи-квадрат
    chi2_crit = chi2.ppf(1 - alpha, df)

    # Решение гипотезы
    decision = chi2_stat > chi2_crit

    return decision, chi2_stat, chi2_crit, k

def chi_criterion_uniform_range(samples, alpha_levels, tetha=48):
    # Определим правила разбиения
    binning_rules = ['Стерджес', 'Скотт', 'Фридмена-Дайкозиса']
    modes = ['uniform']

    # Создаем пустой список для хранения строк данных
    rows = []

    for n, sample_list in samples.items():
        for i, sample in enumerate(sample_list, 1):
            sample_label = f"Выборка размером {n} номер {i}"
            # Рассчитаем k по разным правилам
            k_sturge, k_scott, k_fd = intervals_calc(sample)
            k_dict = {
                'Стерджес': k_sturge,
                'Скотт': k_scott,
                'Фридмена-Дайкозиса': k_fd,
            }
            for alpha in alpha_levels:
                # Инициализируем счетчик решений "Отвергнута"
                reject_count = 0
                # Инициализируем словарь для хранения результатов всех комбинаций
                test_results = {}
                for rule, k in k_dict.items():
                    for mode in modes:
                        decision, chi2_stat, chi2_crit, k_final = chi_square_test_uniform_range(sample, 
                                                                                                tetha=tetha, 
                                                                                                alpha=alpha, 
                                                                                                k_initial=k,
                                                                                                min_expected=5
                                                                                               )
                        # Формируем ключи
                        k_key = f'Количество интервалов {rule} mode = \'{mode}\''
                        chi2_key = f'Xи-квадрат {rule} mode = \'{mode}\''
                        crit_key = f'Критическое значение {rule} mode = \'{mode}\''
                        decision_key = f'Решение {rule} mode = \'{mode}\''

                        # Сохраняем результаты
                        test_results[k_key] = k_final
                        test_results[chi2_key] = round(chi2_stat, 4)
                        test_results[crit_key] = round(chi2_crit, 4)
                        test_results[decision_key] = 'Отвергнута' if decision else 'Принята'

                        # Считаем количество отказов
                        if decision:
                            reject_count += 1

                # Общее решение: отвергнута, если >=2 из 3 решений
                overall_decision = 'Отвергнута' if reject_count >=2 else 'Принята'

                # Создаем строку для DataFrame
                row = {
                    'Выборка': sample_label,
                    'Уровень значимости': alpha,
                    'Решение': overall_decision
                }
                # Добавляем результаты всех комбинаций
                row.update(test_results)
                # Добавляем строку в список
                rows.append(row)

    # Создаем DataFrame из списка строк
    df = pd.DataFrame(rows)

    # Определим порядок колонок
    columns_order = ['Выборка', 'Уровень значимости']

    # Добавляем для каждого правила и режима:
    # - Количество интервалов
    # - Xи-квадрат
    # - Критическое значение
    # - Решение
    binning_rules = ['Стерджес', 'Скотт', 'Фридмена-Дайкозиса']
    modes = ['uniform']
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Количество интервалов {rule} mode = \'{mode}\'')
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Xи-квадрат {rule} mode = \'{mode}\'')
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Критическое значение {rule} mode = \'{mode}\'')
    for rule in binning_rules:
        for mode in modes:
            columns_order.append(f'Решение {rule} mode = \'{mode}\'')

    # Добавляем общее решение
    columns_order.append('Решение')

    # Проверим, что все колонки есть
    existing_columns = df.columns.tolist()
    missing_columns = [col for col in columns_order if col not in existing_columns]
    if missing_columns:
        print(f"Внимание: следующие колонки отсутствуют и не будут добавлены в порядок: {missing_columns}")

    # Реорганизуем колонки, добавляя только те, что есть
    columns_order_filtered = [col for col in columns_order if col in existing_columns]
    df = df[columns_order_filtered]

    return df
```

### Ответы на вопросы
Пусть нам дана выборка $X$ из неизвестного распределения $F_X ∈ 𝓕$, где $𝓕$ - совокупность всех доступных распределений выборки $X$. Выделим подмножество $𝓕_0 ∈ 𝓕$ \
Тогда будем называть нулевой гипотезой $H_0$ следующее утверждение:
$$ F_X ∈ 𝓕_0 $$
А альтернативной гипотезой $H_1$ следующее:
$$ F_X ∈ 𝓕_1, \; 𝓕_1 = 𝓕  ∖ 𝓕_0 $$
Если множество $𝓕_i$ состоит из одного распределения, то будем называть $H_i$ **простой** гипотезой, иначе **сложной** гипотезой \
Будем называть статистическим критерием правило, по которому каждой реализации выборки ставится в соответствие решение: принимаем гипотезу $H_0$ или отвергаем ее (то есть принимаем гипотезу $H_1$) \
Если $X$ - выборка, то $𝔛$ - выборочное пространство, состоящее из всех возможных значений реализации выборки $\bar{x}$. \
Говоря о критерии мы подразумеваем некоторое правило, согласно которому по каждому $x ∈ 𝔛$ мы определяем истинность гипотезы $H_0$. Зададим статистический критерий через следующие 2 множества: \
$𝔛_0 ⊂ 𝔛 ⇒ $ принимается $H_0$, $\; 𝔛_0$ - область принятия гипотезы \
$𝔛_1 ⊂ 𝔛 ∖ 𝔛_0 ⇒ $ принимается $H_1$, $\; 𝔛_1$ - критическая область \
В дальнейшем будем отождествлять понятие статистического критерия и критической области \
Хочется, чтобы вероятность отвергнуть верную нулевую гипотезу $P(\: X ∈ 𝔛_1 \: | \: H_0)$ была как можно меньше. Зафиксируем уровень значимости критерия $α$ так, чтобы: \
$$ P(\: X ∈ 𝔛_1 \: | \: H_0) \leq α$$
Если это неравенство выполнено, то будем говорить, что критерий $𝔛_1$ имеет уровень значимости $α$ и обозначать его $𝔛_{1, α}$ \
Введем понятия ошибок критерия для случая, когда $H_0$ и $H_1$ - простые гипотезы: \
Ошибка 1-го рода соответствует ситуации, когда отвергнута верная нулевая гипотеза: \
$$ α = P(\: X ∈ 𝔛_1 \: | \: H_0)$$
Ошибка 2-го рода соответствует ситуации, когда принята неверная нулевая гипотеза: \
$$ β = P(\: X ∈ 𝔛_0 \: | \: H_1)$$
Функцией мощности критерия W называетя функционал на множестве допустимых распредлений 𝓕 и выборке X. \
$$
W(F_x) = W(F_x; 𝔛_{1,\alpha}) = P(X \in 𝔛_{1, \alpha} | F_x)
$$
$P(X \in 𝔛_{1, \alpha} | F_x)$ - вероятность попасть в $𝔛_{1, \alpha}$, если $F_x$ - итинное распределение \
В нашем случае. Рассмотрим две простые гипотезы $H_0$, $H_1$. Иными словами, считаем, что допустимыми распределениями случайной величины $\xi$ являются два заранее заданных распределения $F_0(x)$ и $F_1(x)$. Необходимо по выборке $X = (X_1, \dots, X_n)$ из распределения $H_0$, $H_1$
$$H_0: \{ \theta = \theta_0\}$$
$$ H_1: \{ \theta = \theta_1 \}$$

### Вычисление функции отношения правдоподобия.
Функция, имеющая вид:
$$
l(\bar{x}) = \frac{L(\bar{x}, \theta_1)}{L(\bar{x}, \theta_0)} = \frac{\prod_{i=1}^n f_1(x_i)}{\prod_{i=1}^n f_0(x_i)}
$$

называется функцией отношения правдоподобия.

#### Дискретное распределение
$$
l(\bar{x}) = \frac{L(\bar{x}, \theta_1)}{L(\bar{x}, \theta_0)} = \frac{\frac{1}{\theta_1 ^ n} Ind(X_{(n)} <= \theta_1)}{\frac{1}{\theta_0 ^ n} Ind(X_{(n)} <= \theta_0)} = (\frac{\theta_0}{\theta_1})^n \frac{Ind(X_{(n)} <= \theta_1)}{Ind(X_{(n)} <= \theta_0)}
$$

#### Непрерывное распределение

$$
l(\bar{x}) = \frac{L(\bar{x}, \theta_1)}{L(\bar{x}, \theta_0)} = \frac{\frac{1}{θ_1^n \sqrt{(2\pi)^n}} \cdot \exp \{-\frac{\sum_{i=1}^n(X_i-\mu)^2}{2θ_1^2} \}}{\frac{1}{θ_0^n \sqrt{(2\pi)^n}} \cdot \exp \{-\frac{\sum_{i=1}^n(X_i-\mu)^2}{2θ_0^2} \}} = (\frac{\theta_0}{\theta_1})^n \exp \{- \frac{\theta_0^2 - \theta_1^2 }{2 \theta_0^2 \theta_1^2} \sum_{i=1}^n (X_i - \mu)^2\}
$$
### Вычисление критической области
Конечно, хорошо, что мы посчитали отношение правдаподобия, но что это нам дает. Мы хотим построить наиболее лучший критерий $𝔛_{1\alpha}^*$, что сводится к максимизации по $𝔛_{1\alpha}$ мощности $W(\theta_1; 𝔛_{1\alpha}) <= \alpha$ \
Наша статистика $l(x) = \frac{L(X; \theta_1)}{X, \theta_0}$ задается критической областью $𝔛_{1\alpha}^* = \{x = (x_1, \dots, x_n): l(x) >= c_\alpha \}$

**Теорема**
При сделанных предположениях критическая область $𝔛_{1\alpha}^*$  задает наиболее мощный критерий для гипотезы $H_0 : F_\xi(x) = F_0(x)$ относительно альтернативы $H_1 : F_\xi(x) = F_1(x)$ среди критериев уровня значимости $\alpha$

#### Непрерывное распределение

$$l(x) >= c$$
$$(\frac{\theta_0}{\theta_1})^n \exp \{- \frac{\theta_0^2 - \theta_1^2 }{2 \theta_0^2 \theta_1^2} \sum_{i=1}^n (X_i - \mu)^2\} >= c$$
Обозначим $T = \sum_{i=1}^n (X_i - \mu)^2$ 

Получается, что у нас 2 случая:
1. $\theta_0 >= \theta_1$ => $T >= \frac{2\theta_1^2 \theta_0^2}{(\theta_0^2 - \theta_1^2)}(nln{\theta_1} - nln{\theta_0} + ln(c)) = t$
2. $\theta_0 <= \theta_1$ => $T <= \frac{2\theta_1^2 \theta_0^2}{(\theta_0^2 - \theta_1^2)}(nln{\theta_1} - nln{\theta_0} + ln(c)) = t$

Мы знаем, что $\frac{T}{\theta} \sim \chi^2_n$
Пусть $X = (X_1, \dots, X_n)$ — выборка из распределения $N(\mu, \theta^2)$ (μ известно), о котором выдвинуты две простые гипотезы $H_0: \theta = \theta_0$ и $H_1: \theta = \theta_1$, где $\theta_1 < \theta_0$. 

и неравенство $l(X) >= c$ эквивалентно неравенству $T \leq t$, поэтому при уровне значимости $\alpha$ критическая граница $t = t_\alpha$ определяется из условия:

$$
\alpha = P[T \leq t_\alpha | H_0] = F_n\left(\frac{t_\alpha}{\theta_0^2}\right),
$$

где $F_n(t)$ — функция распределения закона $\chi^2(n)$. Отсюда $t_\alpha / \theta_0^2 = \chi^2_{\alpha, n}$. Искомый критерий имеет вид:

$$
𝔛_{1\alpha}^{-*} = \{T \leq \theta_0^2 \chi^2_{\alpha, n} \}.
$$

Его мощность:

$$
W(\theta_1; 𝔛_{1\alpha}^{-*} ) = P[T \leq \theta_0^2 \chi^2_{\alpha, n} | H_1] = F_n\left(\chi^2_{\alpha, n} \frac{\theta_0^2}{\theta_1^2}\right).
$$

$$
\beta = 1 - W(\theta_1; 𝔛_{1\alpha}^{-*} ) = 1 - F_n\left(\chi^2_{\alpha, n} \frac{\theta_0^2}{\theta_1^2}\right)
$$

Аналогично находим, что при $\theta_1 > \theta_0$ критерий имеет вид:
$$
\alpha = P[T >= t_\alpha | H_0] = 1 -  P[T <= t_\alpha | H_0] = 1-F_n\left(\frac{t_\alpha}{\theta_0^2}\right)
$$
$$
𝔛_{1\alpha}^{+*}  = \{T \geq \theta_0^2 \chi^2_{1-\alpha, n} \},
$$

и имеет мощность:

$$
W(\theta_1; 𝔛_{1\alpha}^{+*}) = 1 - F_n\left(\chi^2_{1-\alpha, n} \frac{\theta_0^2}{\theta_1^2}\right).
$$

$$
\beta = F_n\left(\chi^2_{1-\alpha, n} \frac{\theta_0^2}{\theta_1^2}\right),
$$

Вычисление минимального необходимого количества материала при фиксации минимального возможного значения ошибок первого и второго рода.

В силу ЦПТ, при большом числе степеней свободы распределение случайной величины $ Y \sim \chi^2(k) $ может быть приближено нормальным $ Y \approx N(k, 2k) $. Более точно:

$$
\frac{Y - k}{\sqrt{2k}} \to N(0, 1) \, \text{по распределению при} \, k \to \infty.
$$

В (1) случае у нас система ($z_\alpha$ квантиль стандартного нормального распредления)
$$
\begin{cases}
t_\alpha = \theta_0^2(n + \sqrt{2n} z_\alpha) \\
t_\alpha = \theta_0^2(n + \sqrt{2n} z_{1-\beta})
\end{cases}
$$

Поэтому $n ≈ [\frac{2(\theta_0^2*z_\alpha - \theta_1^2*z_{1-\beta})}{(\theta_0^2 - \theta_1^2)^2}$] \
Во (2) случае аналогично $n ≈ [\frac{2(\theta_0^2*z_{1-\alpha} - \theta_1^2*z_\beta)}{(\theta_0^2 - \theta_1^2)^2}]$

#### Дискретное распределение
$$l(x) >= c$$
$$
l(x) = (\frac{\theta_0}{\theta_1})^n \frac{Ind(X_{(n)} <= \theta_1)}{Ind(X_{(n)} <= \theta_0)} >=c
$$

$$
l(x) = \frac{L(\theta_1; x)}{L(\theta_0; x)} = 
\begin{cases} 
\left(\frac{\theta_0}{\theta_1}\right)^n, & \text{если } X_{(n)} \leq \min(\theta_0, \theta_1), \\
0, & \text{если } \theta_1 < X_{(n)} \leq \theta_0, \\
\infty, & \text{если } \theta_0 < X_{(n)} \leq \theta_1.
\end{cases}
$$

**Cлучай 1** $\theta_1 < \theta_0$

Если $\theta_1 < \theta_0$, критическая область:
$$
l(X) \geq c \quad \Leftrightarrow \quad X_{(n)} \leq c,
$$
где $c$ определяется из условия:
$$
P(X_{(n)} \leq c \mid H_0) = \alpha \quad \Rightarrow \quad \left(\frac{c}{\theta_0}\right)^n = \alpha \quad \Rightarrow \quad c = \theta_0 \cdot \alpha^{1/n}.
$$

$$
W(\theta_1; 𝔛_{1\alpha}^{-*}) = P(X_{(n)} \leq c \mid H_1) = \left(\frac{c}{\theta_1}\right)^n.
$$
Подставляя $c = \theta_0 \cdot \alpha^{1/n}$:
$$
W(\theta_1; 𝔛_{1\alpha}^{-*}) = \left(\frac{\theta_0 \cdot \alpha^{1/n}}{\theta_1}\right)^n = \left(\frac{\theta_0}{\theta_1}\right)^n \cdot \alpha.
$$

$$
\beta = 1 - \left(\frac{\theta_0}{\theta_1}\right)^n \cdot \alpha.
$$

**Случай 2** $\theta_1 > \theta_0$

Если $\theta_1 > \theta_0$, критическая область:
$$
l(\mathbf{X}) \geq c \quad \Leftrightarrow \quad X_{(n)} \geq c,
$$
где $c$ определяется из условия:
$$
P(X_{(n)} \geq c \mid H_0) = \alpha \quad \Rightarrow \quad 1 - \left(\frac{c-1}{\theta_0}\right)^n = \alpha \quad \Rightarrow \quad c = \theta_0 \cdot (1 - \alpha)^{1/n} + 1.
$$

$$
W(\theta_1; 𝔛_{1\alpha}^{+*}) = P(X_{(n)} \geq c \mid H_1) = 1 - \left(\frac{c-1}{\theta_1}\right)^n.
$$
Подставляя $c = \theta_0 \cdot (1 - \alpha)^{1/n} + 1$:
$$
W(\theta_1; 𝔛_{1\alpha}^{+*}) = 1 - \left(\frac{\theta_0 \cdot (1 - \alpha)^{1/n}}{\theta_1}\right)^n = 1 - \left(\frac{\theta_0}{\theta_1}\right)^n \cdot (1 - \alpha).
$$

$$
\beta = \left(\frac{\theta_0}{\theta_1}\right)^n \cdot (1 - \alpha)
$$

Вычисление минимального необходимого количества материала при фиксации минимального возможного значения ошибок первого и второго рода.

**Cлучай 1** $\theta_1 < \theta_0$
$$
\left(\frac{\theta_0}{\theta_1}\right)^n \cdot \alpha \geq 1 - \beta \quad \Rightarrow \quad n \geq \frac{\ln((1 - \beta)/\alpha)}{\ln(\theta_0 / \theta_1)}.
$$

**Случай 2** $\theta_1 > \theta_0$
$$
\left(\frac{\theta_0}{\theta_1}\right)^n \cdot (1 - \alpha) \leq \beta \quad \Rightarrow \quad n \geq \frac{\ln(\beta / (1 - \alpha))}{\ln(\theta_0 / \theta_1)}.
$$

# Дополнительные слова про непрерывное распредление
Для  k > 30 : Нормальное приближение становится приемлемым, и с увеличением  k  соответствие только улучшается.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def perform_ks_test(k, n=1000, random_seed=42, plot=False):
    np.random.seed(random_seed)

    # Генерация выборки из chi-squared распределения
    Y = np.random.chisquare(df=k, size=n)

    # Стандартизация выборки
    Z = (Y - k) / np.sqrt(2 * k)

    # Выполнение K-S теста
    ks_statistic, p_value = stats.kstest(Z, 'norm')

    print(f"Степени свободы (k): {k}")
    print(f"Статистика K-S теста: {ks_statistic:.5f}")
    print(f"p-значение: {p_value:.5f}")

    if plot:
        # Построение ECDF
        sorted_Z = np.sort(Z)
        ecdf = np.arange(1, n+1) / n

        # Теоретическая CDF стандартного нормального распределения
        theoretical_cdf = stats.norm.cdf(sorted_Z)

        plt.figure(figsize=(8, 6))
        plt.step(sorted_Z, ecdf, label='Эмпирическая CDF', where='post')
        plt.plot(sorted_Z, theoretical_cdf, 'r--', label='Теоретическая CDF N(0,1)')
        plt.xlabel('Z')
        plt.ylabel('CDF')
        plt.title(f'K-S тест для k={k}')
        plt.legend()
        plt.grid(True)
        plt.show()

    return ks_statistic, p_value

# Параметры
degrees_of_freedom = [5, 10, 30, 50, 100, 1000]  # Разные значения k
sample_size = 1000  # Размер выборки
seed = 42  # Для воспроизводимости

# Цикл по разным k и выполнение теста
for k in degrees_of_freedom:
    perform_ks_test(k=k, n=sample_size, random_seed=seed, plot=False)
    print("-" * 50)

for k in [5, 30, 1000]:
    perform_ks_test(k=k, n=sample_size, random_seed=seed, plot=True)
```
